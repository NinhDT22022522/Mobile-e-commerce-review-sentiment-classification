```python
def convert_labels_to_columns(data):
    # Tạo DataFrame mới để lưu trữ kết quả
    result_df = pd.DataFrame()
    
    # Dùng apply() để duyệt qua từng dòng và tạo DataFrame từ mỗi dòng

    dfs = data.apply(lambda row: pd.Series({
        'Pin': 1 if 'BATTERY' in row['label'] else 0,
        'Service': 1 if 'SER&ACC' in row['label'] else 0,
        'General': 1 if 'GENERAL' in row['label'] else 0,
        'Others': 1 if any(label in row['label'] for label in ['OTHERS', 'PERFORMANCE', 'PRICE', 'DESIGN', 'CAMERA', 'SCREEN']) else 0,

        'SPin': 1 if 'BATTERY#Positive' in row['label'] else (-1 if 'BATTERY#Negative' in row['label'] else (0 if 'BATTERY' in row['label'] else 2)),
        'SSer': 1 if 'SER&ACC#Positive' in row['label'] else (-1 if 'SER&ACC#Negative' in row['label'] else (0 if 'SER&ACC' in row['label'] else 2)),
        'SGeneral': 1 if 'GENERAL#Positive' in row['label'] else (-1 if 'GENERAL#Negative' in row['label'] else (0 if 'GENERAL' in row['label'] else 2)),
        'SOth': 1 if any(label in row['label'] for label in ['PERFORMANCE#Positive', 'PRICE#Positive', 'DESIGN#Positive', 'CAMERA#Positive', 'SCREEN#Positive']) else 
            (-1 if any(label in row['label'] for label in ['PERFORMANCE#Negative', 'PRICE#Negative', 'DESIGN#Negative', 'CAMERA#Negative', 'SCREEN#Negative']) else 
            (0 if any(label in row['label'] for label in ['OTHERS', 'PERFORMANCE', 'PRICE', 'DESIGN', 'CAMERA', 'SCREEN']) else 2))

    }), axis=1)

    # Kết hợp các DataFrame tạo từ mỗi dòng thành một DataFrame duy nhất
    result_df = pd.concat([result_df, dfs], ignore_index=True)

    return result_df
result = convert_labels_to_columns(df)
result
```
