```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('labeled_filtered_data.csv')
df.shape
```

```python
import re
from collections import Counter
commit_messages = ' '.join(df['Comments'])
commit_messages = commit_messages.lower()
words = re.findall(r'\b\w+\b', commit_messages)
word_counts = Counter(words)
most_common_words = word_counts.most_common(100)

word_counts['pin']
```

```python
aspect = {'Pin': 0, 'Camera': 0, 'Nhiệt độ': 0, 'Màn hình': 0, 'Nhân viên': 0}
aspect['Pin'] = word_counts['pin'] + word_counts['battery']
aspect['Camera']= word_counts['camera'] + word_counts['cam'] + word_counts['ảnh']
aspect['Nhiệt độ'] = word_counts['mát'] + word_counts['nóng'] + word_counts['lạnh'] + word_counts['nhiệt']
aspect['Màn hình'] = word_counts['màn hình'] + word_counts['màn']
aspect['Nhân viên'] = word_counts['nhân viên'] + word_counts['nv'] 
```
